## 2.Propose a Generative Model

### Suggesting a generative model

Procedure to be followed:<br/>
    -Generate point between 0,$2\pi$<br/>
        ->This can be done by generating uniformly distributed variable between 0 and $2\pi$<br/>
        ->Also we can generate normally distributed random variable with mean 0 and variance $2\pi$<br/>
    -Determine the mean and variance for $X_1$ and $X_2$<br/>
    -Calculate $X_1$ and $X_2$<br/>
        ->For $X_1$ take the cosine of the generated variable, multiply with certain standard deviation $\sigma_1$ and sum with a certain mean $\mu_1$<br/>
        ->For $X_2$ take the sine of the generated variable, multiply with certain standard deviation $\sigma_2$ and sum with a certain mean $\mu_2$<br/>
    -Add Noise to the points<br/>
        ->The data has more noise in the x axis then y axis, therefore we need to specify different errors for the both axices.<br/>

$$U \tilde\ Uniform(0,2\pi)$$
$$X_1 = \sigma_1*cos(U)+\mu_1+\epsilon_1$$
$$X_2 = \sigma_2*sin(U)+\mu_2+\epsilon_2$$
$$\epsilon_1 \tilde\ N(0,\psi_1)$$
$$\epsilon_2 \tilde\ N(0,\psi_2)$$




I expect model to represent the data in the best way. Therefore I try to find such $\sigma_1$ and $\sigma_2$ to make the variance of $X_1$ and $X_2$ close to variances of $X$ and $Y$ coming from the original data respectively. Also I try to find $\mu_1$ and $\mu_2$ that make the mean of the $X_1$ and $X_2$ close to $X$ and $Y$. 

$$Var[X_1]=\sigma_1^2*Var[cos(U)]+Var[\epsilon_1]$$
$$Var[cos(U)]=E[cos(U)^2]-E[cos(U)]^2=\int\limits_{t=0}^{2\pi}cos^2(t)*1/2\pi*dt-0$$
$$Var[cos(U)]=1/2-0=1/2$$
$$Var[X_1]=\sigma_1^2/2+\psi_1$$

$$Var[X_2]=\sigma_2^2*Var[cos(U)]+Var[\epsilon_2]$$
$$Var[cos(U)]=E[cos(U)^2]-E[cos(U)]^2=\int\limits_{t=0}^{2\pi}cos^2(t)*1/2\pi*dt-0$$
$$Var[cos(U)]=1/2-0=1/2$$
$$Var[X_2]=\sigma_2^2/2+\psi_2$$

$$RESULT:$$
$$Var[X_1]=\sigma_1^2/2+\psi_1\approx0.5$$
$$Var[X_2]=\sigma_2^2/2+\psi_2\approx5$$

$$E[X_1]=\sigma_1*E[U]+\mu_1=\sigma_1*0+\mu_1=\mu_1\approx5$$
$$E[X_2]=\sigma_2*E[U]+\mu_2=\sigma_2*0+\mu_2=\mu_2\approx5$$

It is desired to make errors as small as possible. However it can be observed from the data that the error of $X$ is much higher than the error of $Y$. From the equations above following inequalites should be satisfied.
$$\sigma_1^2\leq1$$
$$\sigma_2^2\leq10$$

As a result of some trials and observations, following parameters are chosen:
$$\sigma_1^2=1$$
$$\sigma_2^2=9$$

Consequently, errors are chosen as:
$$\epsilon_1=0.1$$
$$\epsilon_2=0.01$$

Selected Model:
$$U \tilde\ Uniform(0,2\pi)$$
$$X_1 = 1*cos(U)+5+\epsilon_1$$
$$X_2 = 3*sin(U)+5+\epsilon_2$$
$$\epsilon_1 \tilde\ N(0,0.15)$$
$$\epsilon_2 \tilde\ N(0,0.015)$$


## 3.Implement Generative Model and Sample Data From Model 

### Sampling from the model without error terms

```python
n1=1 ##Number of dataset desired to generate
if n1 is 1:
    u = np.random.uniform(0,2*np.pi,size=100) #Generating 100 unifromly distributed RV 
    x1 = np.cos(u)*np.sqrt(1)+5 #Generating X1
    x2 = np.sin(u)*np.sqrt(9)+5 #Generating X2
    generatedData_withoutError=np.column_stack((x1,x2)) #Merging X1 and X2
else:
    generatedData_withoutError=list()
    for i in range(0,n1):
        u = np.random.uniform(0,2*np.pi,size=100)
        x1 = np.cos(u)*np.sqrt(1)+5
        x2 = np.sin(u)*np.sqrt(9)+5
        generatedData_withoutError.append(np.column_stack((x1,x2)))
print generatedData_withoutError
```

### Sampling From Generated Data with Errors

```python
n2=1 ##Number of dataset desired to generate
if n2 is 1:
    u = np.random.uniform(0,2*np.pi,size=100)
    x1 = np.cos(u)*np.sqrt(1)+5+np.random.normal(0,0.15,100)
    x2 = np.sin(u)*np.sqrt(9)+5+np.random.normal(0,0.015,100)
    generatedData_withError=np.column_stack((x1,x2))
else:
    generatedData_withError=list()
    for i in range(0,n2):
        u = np.random.uniform(0,2*np.pi,size=100)
        x1 = np.cos(u)*np.sqrt(1)+5+np.random.normal(0,0.15,100)
        x2 = np.sin(u)*np.sqrt(9)+5+np.random.normal(0,0.015,100)
        generatedData_withError.append(np.column_stack((x1,x2)))
print generatedData_withError
```

## 5.Assessing the similarity of two dataset

In order to compare two datasets and assess the similarity of two datasets statistically, we can use couple of approaches:<br/>
    Kolmogorov-Smirnov Test For 2 Samples (For 2-D Data) can yield some results which tells us whether these two datasets are coming from the same distribution (generating function). Since there is not readily available 2-D version of this test. We can use Kolmogorov-Smirnov Test For 2 Samples for each axis.<br/>
    Another approach would be calculating the MSE's for the generating data and comparing it with the alternative models to understand which one fits better.<br/>
    Similar to the Kolmogorov-Smirnov Test we can use Chi-Square Goodness of Fit test to again assess whether they come from the same distribution or not.<br/>
    In order to check whether the means are equal or not, we can use t-test and similarly in order to check whether the variances are equal or not we can use f test.<br/>
    We can use q-q plot.<br/>
    Lastly we can look at the correlatin between the datasets.

### Kolmogorov Smirnov Test 

```python
##Can be used when single data is generated, otherwise it selects the first data
if n2 is 1:
    print 'Case with no Error:\n'
    print stats.ks_2samp(data[:,0], generatedData_withoutError[:,0])
    print stats.ks_2samp(data[:,1], generatedData_withoutError[:,1]),"\n"
    print 'Case with Error:\n'
    print stats.ks_2samp(data[:,0], generatedData_withError[:,0])
    print stats.ks_2samp(data[:,1], generatedData_withError[:,1])
else:
    print 'Case with no Error:\n'
    print stats.ks_2samp(data[:,0], generatedData_withoutError[0][:,0])
    print stats.ks_2samp(data[:,1], generatedData_withoutError[0][:,1]),"\n"
    print 'Case with Error:\n'
    print stats.ks_2samp(data[:,0], generatedData_withError[0][:,0])
    print stats.ks_2samp(data[:,1], generatedData_withError[0][:,1])
```

From the result above we can see that the p-values are high for the both cases. Therefore we can conclude that the simulated dataset from the generating function is similar to the given data.
